```python
import cv2 
import random
from matplotlib import pyplot as plt 
```

```python
def add_noise(img):
    row, col = img.shape # to get the dimensions of the image 
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row-1)
        xcord = random.randint(0, col-1)
        img[ycord][xcord] = 255
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row-1)
        xcord = random.randint(0, col-1)
        img[ycord][xcord] = 0
        
    return img
```

```python
image = cv2.imread('Car.jpg')
plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
#adding the noise to the picture Code to create Salt and pepper effect on the image
img = cv2.imread('Car.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('Carwithnoise.jpg', add_noise(img))
noise_image = cv2.imread('Carwithnoise.jpg')
plt.imshow(noise_image)
plt.show()
```

```python
#enhancing the image 
median = cv2.medianBlur(noise_image, 5)
plt.imshow(median)
plt.show()
```
