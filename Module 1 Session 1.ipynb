```python
import cv2 #this will be useful in performing image flipping operation
from matplotlib import pyplot as plt  #used to display the images in jupyter
```

```python
img = cv2.imread('Opencv.png')
name = 'OpenCV Python Image'
plt.title(name)
plt.imshow(img)
```

```python
#Fliping the image horizontally 
horizontal_flip = cv2.flip(img, 1)  #image code 1 flips the image around y axis. 
name="OpenCV Python Image Horizontally"
plt.title(name)
plt.imshow(horizontal_flip)
```

```python
#Fliping the image vertically
vertical_flip = cv2.flip(img, 0)
name = 'Opencv Flipping the image vertically'
plt.title(name)
plt.imshow(vertical_flip)
```

```python
vertical_horizontal_flip = cv2.flip(img, -1)
name = 'OpenCV Python Flip Image'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
