```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Machine learning 
from sklearn_extra.cluster import KMedoids

import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating dummy image data
#Dummy the image
data = np.array([[7,8], [9,10], [11,5], [4,9], [7,5], [2,3], [4,5]])
```

```python
#Visualizing the dataset
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[:, 0], y = data[:, 1], marker='o', s=100)
plt.title("Dummy Image data points")
plt.show()
```

```python
#Initializing centers Manually
centers = [[4,5], [9,10]]
```

```python
#Creating a KMedoids model
Kmedoids = KMedoids(n_clusters=2)
Kmedoids.fit(data)

#Assigning the labels
labels = Kmedoids.labels_
```

```python
#Visualizing the cluster formed
#Extraction of unique labels
unique_labels = set(labels)

#color codes
colors_plot = [
    plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))
]

#Set data point feature X and feature Y
xy = data

#Plotting feature X and feature Y for each cluster labels
plt.figure(figsize=(10, 8))
sns.scatterplot(x = xy[:, 0], y = xy[:, 1], marker='+', hue=labels, s=100)

#Annonate the centroid
sns.scatterplot(Kmedoids.cluster_centers_[:, 0], Kmedoids.cluster_centers_[:, 1], marker='o', color="red", label='Centroids', s=100)
plt.title("KMedoids clustering on dummy data", fontsize=14)
plt.legend()
plt.show()
```
