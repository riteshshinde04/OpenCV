```python
#Linear Transformation - converts an image into its inverse 
import cv2 
import numpy as np
import matplotlib.pyplot as plt
im = cv2.imread('Car.jpg', 0)
imginv = 255 - im

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize=18)
plt.imshow(im,cmap='gray') #displays the grayscale image
ax = fig.add_subplot(122)
plt.imshow(imginv,cmap='gray')
ax.set_title('Inverse Transfrom', fontsize=18)
```
