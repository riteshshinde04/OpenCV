```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np 

im = cv2.imread('Car.jpg', 0)
im1 = im/255.0
im_power_law_transformation = cv2.pow(im1,1.8)
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121) 
plt.title('Original', fontsize=18)
plt.imshow(im,cmap='gray')
ax = fig.add_subplot(122) #used to add figure in the image as a part of subplot of the image 
plt.imshow(im_power_law_transformation,cmap='gray')
ax.set_title('Power Law Transformation', fontsize=18)
```
