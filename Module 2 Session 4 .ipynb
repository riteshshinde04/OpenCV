```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Reading the image and setting up the kernel
img = cv2.imread('Car.jpg')
kernel = np.ones((5,5),np.uint8)
plt.imshow(img)
```

```python
#Erosion
erosion = cv2.erode(img,kernel)
plt.imshow(erosion)
```

```python
#dilation
dilation = cv2.dilate(img,kernel)
plt.imshow(dilation)
```
